```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load data
file_name = 'cars93.csv'
df = pd.read_csv(file_name)

# Displays first few rows of the dataframe
df.head()
df.columns
```

### 1. Boxplot

```python
# Boxplot highest rev per mile

# Filtering the DataFrame to include only rows with the selected manufacturers
selected_manufacturers = df[df['Manufacturer'].isin(['Audi', 'Suzuki', 'Hyundai', 'Toyota'])]

# Plotting the boxplot
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=selected_manufacturers)
# Finding out who has highest rev per miles
highest_rev_manufacturer = selected_manufacturers.groupby('Manufacturer')['Rev.per.mile'].mean().idxmax()

# Labeling and titling with an f-string to include the manufacturer with the highest revs per mile
plt.xlabel('Manufacturer')  
plt.ylabel('Revs Per Mile')
plt.title(f'Average Revs Per Mile by Manufacturer\nHighest: {highest_rev_manufacturer}')  

plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
The answer is Suzuki

### 2. Histogram

```python
# Histogram of MPG

# Determine if it's generally better to drive in city or highway based on average MPG
better_driving = "Highway" if df['MPG.highway'].mean() > df['MPG.city'].mean() else "City"

# Plotting the histogram for MPG.highway and MPG.city
sns.histplot(data=df[['MPG.highway','MPG.city']], bins=12, palette=['blue', 'orange'])

# Labeling the plot correctly with an f-string to include where it is better to drive
plt.xlabel('Miles Per Gallon (MPG)')  
plt.ylabel('Frequency')               
plt.title(f'Distribution of Miles Per Gallon (MPG) for Highway and City Driving\nGenerally Better to Drive: {better_driving}') 

# Show the plot
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
It is more fuel efficient to drive in the Highway

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Turn.circle', y='Wheelbase', data=df)

# Set labels for axes
plt.xlabel('Turn.circle')
plt.ylabel('Wheelbase')    

# Set legend
plt.legend(['Wheelbase', 'Turn.circle'])

# Show the plot
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
As the wheelbase increases, the turn circle tends to increase as well. This suggests that vehicles with longer wheelbases typically require a larger turning radius to complete a full circle.

### 4. Barplot

```python
# Barplot for mean horsepower

# Group the DataFrame by car type and calculate the mean horsepower
car_size_hp = df.groupby('Type')['Horsepower'].mean().reset_index()

# Checking the dataframe
# print(car_size_hp)

# Plot the barplot with proper labeling and different colors for each car type
sns.barplot(x='Type', y='Horsepower', data=car_size_hp, hue='Type', palette='tab10', legend=False)

plt.xlabel('Type')  
plt.ylabel('Horsepower')               
plt.title('Average Horsepower Of Each Car Types') 
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes the as the car get bigger the more horsepower it has
