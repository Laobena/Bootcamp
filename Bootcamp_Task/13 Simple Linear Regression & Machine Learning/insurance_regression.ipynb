```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

```

```python
# Import the dataset and take a look at the beginning of the data.
df = pd.read_csv("insurance.csv")
df.head()
```

```python
#Checking for missing data
df.isnull().sum()
```

```python
# Changing df to suit my requirements.
df =df[['age','charges']]


# Using Scatterplot to Visualise Data
sns.scatterplot(data=df, x='charges', y='age')
plt.xlabel('Charges')
plt.ylabel('Age')
plt.show()

```

```python
# Preparing the data for simple linear regression
X = df['charges'].values.reshape(-1, 1)
y = df['age'].values
```

```python
# Creating and fitting the linear regression model
simple_model = LinearRegression()
simple_model.fit(X, y)
```

```python
# Making predictions
y_pred = simple_model.predict(X)
```

```python
sns.scatterplot(data=df, x='charges', y='age')
plt.plot(X, y_pred, color='red')
plt.xlabel('Charges')
plt.ylabel('Age')
plt.title('Linear Regression')
plt.show()
```
